```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0, 10, 20)
y = x**2
```

```python
plt.plot(x, y, 'r.')
plt.show() # not required in jupyter notebook
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0, 0, 1, 1])
axes2 = fig.add_axes([0.2, 0.7, 0.2, 0.2]) #[from left, from bottom, width of canvas, height of canvas ]
axes3 = fig.add_axes([0.5, 0.2 ,0.4, 0.5])
```

```python
fig = plt.figure()

count = 1
axes = fig.subplots(nrows = 2, ncols = 3)  #indicating nuber of rows and columns of graph
'''Since there are 2 rows and three column it is 2D i.e 2X3 matrix and has 2*3= 6 graph box'''
for ax in axes: # for row
    for current_axes in ax: # for column
        y = x**count
        current_axes.plot(x, y)
        count += 1
        
plt.tight_layout()
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0, 0, 1, 1])

axes1.plot(x, x**2, label='X squared')
axes1.plot(x, x**3, label='X cubed', )

plt.legend(loc = 0)
```
