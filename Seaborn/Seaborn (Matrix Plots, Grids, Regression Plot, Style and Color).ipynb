```python
import seaborn as sns
%matplotlib inline

tips=sns.load_dataset('tips')
flights=sns.load_dataset('flights')
tips.head()
```

```python
flights.head()
```

### Heat Map

Colors value in sort of gradient scale

```python
tc = tips.corr()
tc
```

```python
sns.heatmap(tc)
```

```python
sns.heatmap(tc, annot=True) # passing annot = True arg the daigram will show the corr value
```

```python
sns.heatmap(tc, annot=True, cmap='coolwarm', linewidths=2)
```

```python
fp = flights.pivot_table(index='month', columns='year', values='passengers')
```

```python
sns.heatmap(fp, cmap='coolwarm', linewidths=2)
```

```python
sns.heatmap(fp, linewidths=2)
```

# Cluster Map

It uses hierarichal clustering to produce a clustered version of heat map

```python
sns.clustermap(fp)
```

#### Cluster show the columns and rows that are similar to each other 

```python
sns.clustermap(fp, standard_scale=1) # standard_scale to normalize the scale
```

# Grids

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

 ### Pair Grid

```python
import matplotlib.pyplot as plt
```

```python
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
g = sns.PairGrid(iris)

g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

### Facet Grid

```python
fg=sns.FacetGrid(data=tips, row='time', col='smoker')
fg.map(plt.scatter, 'total_bill', 'tip') #scatter plot will have tip vs total_bill
```

```python
fg=sns.FacetGrid(data=tips, row='time', col='smoker')
fg.map(sns.distplot, 'total_bill') 
```

# Regression Plot

```python
sns.lmplot(x='total_bill', y='tip', data=tips)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o', 'x'], scatter_kws={'s':80})
```

##### Markers is used to indicate different types in data and using scatter_kws change the size of markers

```python
sns.lmplot(x='total_bill', y ='tip', data=tips, col='sex')
```

###### using col instead of hue we get different column for different category

```python
sns.lmplot(x='total_bill', y ='tip', data=tips, col='sex', row='time')
```

```python
sns.lmplot(x='total_bill', y ='tip', data=tips, col='sex', hue='day')
```

```python
sns.lmplot(x='total_bill', y ='tip', data=tips, col='sex', hue='day', aspect=0.8, height=8)
```

###### aspect is used for maintaining aspect ratio 

# Style and Color

```python
sns.set_style('whitegrid')
sns.countplot(x='sex', data=tips)
```

```python
sns.set_style('ticks')
sns.countplot(x='sex', data=tips)
sns.despine() # also can pass agrument for left bottom right top
```

```python
sns.set_style('ticks')
sns.set_context('paper')
sns.countplot(x='sex', data=tips)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', palette='coolwarm')
```
