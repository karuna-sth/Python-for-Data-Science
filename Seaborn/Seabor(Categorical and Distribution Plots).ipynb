```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

# Distribution Plots

```python
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'], kde=False)
```

```python
sns.distplot(tips['total_bill'], kde=False, bins=30)
```

#### Jointplot

```python
sns.jointplot(x='total_bill', y='tip', data=tips)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind = 'hex')
```

more points more darker

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind = 'reg') #mostly used
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind = 'kde')
```

### Pairplot

```python
sns.pairplot(tips) #make different pair and plots data
```

Hue passin column name like sex as it has two catagory f and m

```python
sns.pairplot(tips, hue='sex')
```

```python
sns.pairplot(tips, hue='sex', palette='coolwarm')
```

### Rug plot

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'], kde=False)
```

The difference in these two are  histogram has bims and counts how many dashes were in bin

# Categorial Plots

### Bar Plot

```python
sns.barplot(x='sex', y='total_bill', data=tips)
```

Shows average total_bill for male and female

```python
sns.barplot(x='day', y='total_bill', data=tips)
```

```python
import numpy as np
```

```python
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std) # std of bill per category
```

We can pass any estimator function and plot on that basis

### Count Plot

##### Counts the occurance of certain data in dataset

```python
sns.countplot(x='sex', data=tips)
```

# Box Plot

```python
sns.boxplot(x = 'day', y= 'total_bill', data=tips) # pass  category in x and numeric dat on y
```

#### The points outside the wiskher are outliers

```python
sns.boxplot(x = 'day', y= 'total_bill', data=tips, hue='smoker')
```

Using hue we split box plot by another categorical column
So the box plot split per day and also split by smoker status 

### Violin Plot

```python
sns.violinplot(x='day', y='total_bill', data=tips)
```

Violin plot show more information than box plot

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

### Strip Plot

```python
sns.stripplot(x='day', y='total_bill', data=tips)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=False)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

Split will split plot based of category

### Swarm Plot

```python
sns.swarmplot(x='day', y='total_bill', data=tips)
```

##### Looks as combination of strip and violin plot

```python
sns.swarmplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

##### NOTE: Swarm plot is not suitable for large data sets as the points go way too wide and stack up

##### Better to combine it with violin plot can be used in EDA 

```python
sns.violinplot(x='day', y='total_bill', data=tips)
sns.swarmplot(x='day', y='total_bill', data=tips, color='black')
```

### Factor Plot

```python
sns.factorplot(x='day', y='total_bill', data=tips, kind='violin')
```

##### Factor plot can be used to draw any plot by passing argument kind='bar' or  
##### kind = 'strip'
